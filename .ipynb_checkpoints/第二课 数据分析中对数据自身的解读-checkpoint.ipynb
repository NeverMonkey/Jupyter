### 数据框

```python
import pandas as pd

data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'],
        'year': [2000, 2001, 2002, 2001, 2002],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9]}
df=pd.DataFrame(data)
df
```

```python
df.index=['row1','row2','row3','row4','row5']
df
```

### 标签分组

```python
### 
group1=['group1','group1','group1','group2','group2']
group2=['row'+str(x) for x in range(1,6)]
df.index=pd.MultiIndex.from_arrays([group1,group2], names=['group', 'row'])
df
```

```python
df.mean(level=['group'])
```

### 因子操作

```python
dumm=pd.get_dummies(df['state'])
dumm
```

```python
pd.concat([df,dumm],axis=1)
```

```python
pd.get_dummies(df,columns=['state','year'])
```

### 空值处理

```python
import os

path=u'.\\Data\\2017-03-15'
dt=pd.read_csv(os.path.join(path,u'CSV导出\\march16 08_19_34绍兴北到宁波.csv'))
dt=dt[['Time','LONG','LAT','RSRP','eNB_ID','SCell CellID']]
dt.rename(columns={'SCell CellID':'CI'},inplace = True)
print dt.columns
dt[5:15]
```

```python
dt2=dt.fillna(method='ffill')
dt2=dt2.dropna()
dt2
```

```python
dt3=dt2.copy()
dt3['eNB_ID']=dt3['eNB_ID'].astype('category')
dt3['eNB_ID']=dt3['eNB_ID'].astype('int').astype('category')
dt3['CI']=dt3['CI'].astype('int').astype('category')
print dt3.info()
dt3.head()
```

### 归一化

```python
from sklearn import preprocessing

dt3['Std_RSRP']=preprocessing.StandardScaler().fit_transform(dt3[['RSRP']])
dt3['Nor_RSRP']=preprocessing.MinMaxScaler().fit_transform(dt3[['RSRP']])
print dt3['RSRP'].describe()
print
print dt3['Std_RSRP'].describe()
print
print dt3['Nor_RSRP'].describe()
```

### 箱型图

```python
import numpy as np
spread = np.random.rand(50) * 150
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low), 0)
plt.boxplot(data)
plt.show()
```
